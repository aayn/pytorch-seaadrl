# Plots

```python
import pickle
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
matplotlib.rcParams['figure.figsize'] = [20, 10]
```

## A2C

## CP Attack

```python
with open('logs/pong/plots/t_vs_ar.pkl',  'rb') as pf:
    ar_vs_threshold = pickle.load(pf)
thresholds, ar = zip(*ar_vs_threshold)
```

```python
plt.plot(thresholds, ar, linewidth=5)
plt.xlabel('$\Delta$ (attack threshold)', fontsize=20)
plt.ylabel('Average return per episode', fontsize=20)
plt.title('Atari Pong CP Attack', fontsize=20)
plt.show()
```

## Antagonist Attack

```python
num_attack_steps = [0, 1, 2, 3, 4, 5]
avg_return = [20.25, 20.25, -20.625, -18.71875, -20.5, -20.6875]
```

```python
plt.plot(num_attack_steps, avg_return, linewidth=5)
plt.xlabel('Number of attack steps', fontsize=20)
plt.ylabel('Average return per episode', fontsize=20)
plt.title('Atari Pong Antagonist Attack', fontsize=20)
plt.show()
```
